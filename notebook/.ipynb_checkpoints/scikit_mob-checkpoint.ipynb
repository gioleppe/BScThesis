In this notebook we do some analysis with scikit mob on the geolife dataset. <br>
We start by importing the first three users of the geolife DB and printing their trajectories.

```python
import skmob

users = list(["{0:03}".format(i) for i in range(3)])
print(users)
tdf = skmob.io.file.load_geolife_trajectories("Desktop/Tesi/datasets/Geolife Trajectories 1.3/", \
                                              user_ids=users, filter_kwargs={'max_speed_kmh': 400})
```

Now we print the trajectories on a Folium map

```python
tdf.plot_trajectory(zoom=12, weight=3, opacity=0.9, tiles='Stamen Toner')
```

Let's try with a flow dataframe

```python
from skmob.tessellation import tilers

tessellation = tilers.tiler.get("squared", base_shape="Macau, Macau", meters=500)

print(tessellation)

print(tdf.head())

fdf = tdf.to_flowdataframe(tessellation=tessellation, self_loops=True, remove_na=True)
print(fdf.head)
```
