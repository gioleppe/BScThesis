# Plotting some informative figures on computed coverages
We proceed to visualize our coverage probabilities with heatmaps and meshgrids

```python
import pandas as pd
import yaml 
import numpy as np
import matplotlib.pyplot as plt

with open("conf.yaml") as f:
    conf = yaml.load(f, Loader = yaml.FullLoader)

base_path = conf["base_path"]
data_path = conf["data_path"]
out_path = conf["out_path"]
```

### Getting all our coverage filenames from the data path

```python
import glob, os, pprint as pp

names =[os.path.basename(file) for file in glob.glob(data_path+"*lambda*")]
names.sort()
pp.pprint(names)
print(len(names)) #expected: 9

for name in names:
    print(name.split("_"))
```

## 3D meshgrid and heatmaps
We plot and save meshgrids and heatmaps representing our coverage probabilities

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from scipy.interpolate import griddata

from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
import pandas as pd

plt.rcParams.update({'font.size': 22})

for name in names:
    
    nameparts = name.split("_")
    #basetitle = nameparts[0] + " " + nameparts[1] + ", lambda set to " + nameparts[2]
    
    data = pd.read_csv(data_path + name)
    numcols, numrows = 30, 30

    fig = plt.figure(figsize=(15,8))
    ax = fig.gca(projection='3d')

    xi = np.linspace(data.lon.min(), data.lon.max(), numrows)
    yi = np.linspace(data.lat.min(), data.lat.max(), numcols)

    #print(xi, yi)

    xi, yi = np.meshgrid(xi, yi)

    #print(xi, yi)

    x, y, z = data.lon, data.lat, data.probability

    #print(x, y, z)

    # known points (long, lat)
    points = np.vstack((x,y)).T 
    # known z points
    values = z
    # all the points I want to interpolate
    wanted = (xi, yi)
    # meshgrid interpolated
    zi = griddata(points, values, wanted,method="linear", fill_value = 0) 

    surf = ax.plot_surface(xi, yi, zi, cmap=cm.spring,
                           linewidth=0, antialiased=False,shade=True)

    ax.set_zlim(0, 1)
    #ax.zaxis.set_major_locator(LinearLocator(10))
    #ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

    # Add a color bar which maps values to colors.
    fig.colorbar(surf, shrink=0.3, aspect=5)
    
    #plt.title(basetitle + " - 3D grid")
    
    plt.savefig(out_path + name.replace(".csv", "") + "_3D_grid.png", dpi  = 300)
    plt.show()
    
    
    plt.figure(figsize=(15, 10))
    plt.contourf(xi, yi, zi)
    plt.scatter(data.lon, data.lat, c=data.probability, s=60, vmin=zi.min(), vmax=zi.max(),cmap=cm.summer)
    plt.colorbar()
    #plt.title(basetitle + ' - contour heatmap')

    plt.savefig(out_path + name.replace(".csv", "") + "_heatmap.png", dpi  = 300)
    plt.show()
```
